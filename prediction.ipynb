```python
import pickle
import pandas as pd
from tensorflow.keras.models import load_model
```

```python
#### Variables ####

from CONFIG import features, directories

# Directories
log_dir = directories["log_dir"]
model_dir = directories["model_dir"]

# Features
categorical_variables = features["categorical_variables"]
numerical_variables = features["numerical_variables"]
binary_variables =  features["binary_variables"]
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
# Loading pkl files and model
with open("one_hot_encoder.pkl", "rb") as f:
    one_hot_encoder = pickle.load(f)

with open("min_max_scaler.pkl", "rb") as f:
    min_max_scaler = pickle.load(f)

model = load_model(f"{model_dir}/model_1.keras")
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
categorical_encoded = pd.DataFrame(one_hot_encoder.transform(input_df[categorical_variables]))
numerical_standardized = pd.DataFrame(min_max_scaler.transform(input_df[numerical_variables]))
preprocessed_df = pd.concat([input_df[binary_variables].reset_index(drop=True), categorical_encoded, numerical_standardized], axis=1)
preprocessed_df
```

```python
model.predict(preprocessed_df)
```
